# California County Data: Air Quality (AQI) and Asthma

```python
import pandas as pd
import requests
```

# 1. Asthma

```python
url = "https://data.cdc.gov/resource/i46a-9kgh.json"
params = {
    "$select": "stateabbr,countyname,countyfips,casthma_adjprev",
    "stateabbr": "CA",
    "$limit": 100,
}

r = requests.get(url, params=params)
r.raise_for_status()

asthma_ca = pd.DataFrame(r.json())
```

```python
asthma_ca.head()
```

```python
asthma_ca.info()
```

```python
asthma_ca.isna().sum()
```

## 2. AQI data

How can i get AQI data by county? SEARCH!

```python
aqi = pd.read_csv("../data/raw/annual_aqi_by_county_2023.csv")
```

```python
aqi.head()
```

```python
aqi_ca = aqi[aqi["State"] == "California"]
aqi_ca.head()
```

```python
aqi_ca = aqi_ca[["State", "County", "Median AQI"]]
aqi_ca.head()
```

```python
aqi_ca.info()
```

```python
aqi_ca.isna().sum()
```

## 3. Merge

```python
county_data = pd.merge(
    asthma_ca,
    aqi_ca,
    left_on="countyname",
    right_on="County",
)
county_data = county_data[["stateabbr", "countyname", "countyfips","casthma_adjprev", "Median AQI"]]
county_data.head()
```

```python
county_data = county_data.rename(
    columns={
        "stateabbr": "state",
        "countyname": "county",
        "countyfips": "fips",
        "casthma_adjprev": "asthma",
        "Median AQI": "aqi",
    }
)
county_data.head()
```

```python
county_data.info()
```

## 4. Save

```python
county_data.to_csv("../data/interim/ca_county_aqi_asthma.csv", index=False)
```
